```python
barragem_algun = barragem.iloc[:,0:20]
barragem_algun
```

```python
barragem_algun.dropna(inplace=True)
barragem_algun.info()
```

```python
def latitude_para_decimal(lat):
    partes = lat.split('°')
    graus = float(partes[0])

    partes_minutos = partes[1].split("'")
    minutos = float(partes_minutos[0])

    partes_segundos = partes_minutos[1].split('"')
    segundos = float(partes_segundos[0])

    if graus < 0:
        graus = -1*graus

    # Calculando a latitude decimal
    decimal = graus + (minutos / 60) + (segundos / 3600)
    return -1*decimal
```

```python
barragem_algun['Lat con'] = barragem_algun['Latitude'].apply(latitude_para_decimal)
barragem_algun['Long con'] = barragem_algun['Longitude'].apply(latitude_para_decimal)
```
